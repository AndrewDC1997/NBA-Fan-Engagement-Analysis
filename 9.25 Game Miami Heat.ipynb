```python
text1 = ' '.join(tweet for tweet in df1_t).replace('https', ' ').replace('.co',' ').replace('de ',' ')
text2 = ' '.join(tweet for tweet in df2_t).replace('https', ' ').replace('.co',' ').replace('de ',' ')
text3 = ' '.join(tweet for tweet in df3_t).replace('https', ' ').replace('.co',' ').replace('de ',' ')
text4 = ' '.join(tweet for tweet in df4_t).replace('https', ' ').replace('.co',' ').replace('de ',' ')
text5 = ' '.join(tweet for tweet in df5_t).replace('https', ' ').replace('.co',' ').replace('de ',' ')
text6 = ' '.join(tweet for tweet in df6_t).replace('https', ' ').replace('.co',' ').replace('de ',' ')
```
